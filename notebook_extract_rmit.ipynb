# RMIT Courses
- Retrieves course high level details from rmit website

## Import modules

```python
import requests
import pandas
import bs4
```

## Store base url 

```python
base_url = "https://www.rmit.edu.au/study-with-us/"
```

```python
request = requests.get(base_url)
```

```python
soup = bs4.BeautifulSoup(request.text, "lxml")
```

## Locate and store category links

```python
section = soup.find("section", {"class":"m-feature-tile-3-col"})
```

```python
categories = section.find_all("a")
```

```python
category_links = []

for item in categories:
    category_links.append(item.get("href"))
```

```python
category_links
```

```python
category_links.pop(0)

category_links
```

```python
full_links = []

for link in category_links:
    
    full_links.append(base_url + link.split("/")[2])
    
full_links
```

```python
full_links2 = []

for link in full_links:
    request = requests.get(link)
    soup = bs4.BeautifulSoup(request.text, "lxml")
    
    subcategories = soup.find_all("div", {"class":"col-sm-6 col-xs-12"})
    for sc in subcategories:
        sc_link = sc.find("a").get("href")
        full_links2.append(base_url + sc_link.replace("/study-with-us/",""))
                           
full_links2
```

```python
len(full_links2)
```

## Create new list containing unique links from full_links

```python
full_links2_nodupes = []

for item in full_links2:
    if item not in full_links2_nodupes:
        full_links2_nodupes.append(item)
        
```

```python
len(full_links2_nodupes)
```

```python
full_links2_nodupes = sorted(full_links2_nodupes)
full_links2_nodupes
```

## Process data from each sublink

```python
data_list = []

for index in range(len(full_links2_nodupes)):
    print("START PROCESSING INDEX", index)
    link = full_links2_nodupes[index]
    
    request = requests.get(link)
    text = request.text
    
    data = pandas.read_html(text)
    data[0]["link"] = link
    
    data_list.append(data[0])
    
len(data_list)
```

## Merge dataframes into one

```python
full_data = pandas.concat(data_list, 
                          axis = 0, 
                          sort = True)
```

```python
full_data.head()
```

## Drop index

```python
full_data = full_data.reset_index(drop = True)
```

```python
full_data.head()
```

```python
full_data.shape
```

```python
for index in range(full_data.shape[0]):
    get_location = full_data.loc[index, "Location"]
    if get_location == "Location":
            full_data.drop([index], axis = 0, inplace = True)
```
