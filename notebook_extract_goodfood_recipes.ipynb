# Recipe pull
- https://www.goodfood.com.au

## Import Modules 

```python
import bs4
import requests
import pandas
import datetime
```

## Set text query
- Recipes to pull is limited to 125 per text query

```python
text_query = "tomato"
```

## Set urls 

```python
urls = ["https://www.goodfood.com.au/search?text="+text_query+"&location=&type=recipe&sortBy=relevance",
        "https://www.goodfood.com.au/search?text="+text_query+"&location=&type=recipe&sortBy=relevance&p=2",
        "https://www.goodfood.com.au/search?text="+text_query+"&location=&type=recipe&sortBy=relevance&p=3",
        "https://www.goodfood.com.au/search?text="+text_query+"&location=&type=recipe&sortBy=relevance&p=4",
        "https://www.goodfood.com.au/search?text="+text_query+"&location=&type=recipe&sortBy=relevance&p=5"]
```

## Create Dictionary to store data

```python
recipe_dictionary = {"difficulty":[],
                    "name":[],
                    "time":[],
                    "url":[],
                    "desc":[],
                    "ingredients":[],
                    "method":[]}
```

```python
test = "https://www.goodfood.com.au/search?text="+text_query+"&location=&type=recipe&sortBy=relevance"
```

## Process 

```python
for link in urls:
    
    print("start processing link",link)
    
    request = requests.get(link)
    soup = bs4.BeautifulSoup(request.text, "lxml")
    grid = soup.find("div",{"class":"grid"})
    articles = grid.find_all("article", {"class":"story"})

    for article in articles:

        recipe_difficulty = article.find("h4", {"class":"story__kicker"}).text.strip()
        recipe_name = article.find("h3", {"class":"story__headline"}).text.strip()
        recipe_url = article.find("h3", {"class":"story__headline"}).find("a").get("href").strip()
        cook_time = article.find("li", {"class":"cooktime"}).text.strip()
        recipe_desc = article.find("p").text.strip()

        request = requests.get(recipe_url)
        subsoup = bs4.BeautifulSoup(request.text, "lxml")
        recipe_ingredients = subsoup.find("div", {"class":"recipe__instructions--ingredients"}).text.strip()
        recipe_method = subsoup.find("div", {"class":"recipe__instructions--method"}).text.strip()

        recipe_dictionary["difficulty"].append(recipe_difficulty)
        recipe_dictionary["name"].append(recipe_name)
        recipe_dictionary["time"].append(cook_time)
        recipe_dictionary["url"].append(recipe_url)
        recipe_dictionary["desc"].append(recipe_desc)
        recipe_dictionary["ingredients"].append(recipe_ingredients)
        recipe_dictionary["method"].append(recipe_method)
```
