# Scrape from dictional book store website

## Import modules

```python
import requests
import pandas
import bs4
import datetime

start = datetime.datetime.now()
```

## Dictionary to store results

```python
dictionary = {"title":[],
             "price":[],
             "stock":[],
             "rating":[],
             "url":[],
             "page_number":[],
             "date_extracted":[]}
```

## Create pages list, website has 50 pages

```python
pages = list(range(1,51))
```

## Start Process, take data from every page and store in dictionary

```python


for page in pages:
    
    print("processing page number",page)
    base_url = "http://books.toscrape.com/catalogue/category/books_1/page-" + str(page) +".html"
    
    request = requests.get(base_url)
    soup = bs4.BeautifulSoup(request.text , "lxml")

    book_list = soup.find("ol")
    book_list_items = book_list.find_all("li")

    for item in book_list_items:
        title = item.find("h3").find("a").get("title")
        price = item.find("p", {"class":"price_color"}).text.strip()
        stock = item.find("p", {"class":"instock availability"}).text.strip()
        rating = item.find("p", {"class":"star-rating"}).get("class")[1]

        dictionary["title"].append(title)
        dictionary["price"].append(price)
        dictionary["stock"].append(stock)
        dictionary["rating"].append(rating)
        dictionary["page_number"].append(page)
        dictionary["url"].append(base_url)
        dictionary["date_extracted"].append(start)
        
        
```

## Convert dictionary to pandas dataframe and preview

```python
data = pandas.DataFrame(dictionary)
```
