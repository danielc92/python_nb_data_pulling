# Scrape Quotes from fictional site
http://quotes.toscrape.com/

## Import modules

```python
import pandas
import requests
import datetime
import bs4
```

```python
date = datetime.datetime.now()
```

```python
dictionary = {"tag":[],
             "quote":[],
             "author":[],
             "date_extracted":[]}
```

```python
for page_number in list(range(1,11)):
    
    print("processing page ",page_number)
    url = "http://quotes.toscrape.com/page/" + str(page_number)
    request = requests.get(url)
    soup = bs4.BeautifulSoup(request.text, "lxml")
    quotes = soup.find_all("div",{"class":"quote"})

    for q in quotes:

        try:
            author = q.find("span", {"class":"author"}).text
        except:
            author = None

        quote = q.find("span", {"class":"text"}).text

        tags = []

        tags_resultset = q.find_all("a", {"class":"tag"})

        for tag in tags_resultset:
            tags.append(tag.text)

        for tag in tags:
            dictionary["tag"].append(tag)
            dictionary["quote"].append(quote)
            dictionary["author"].append(author)
            dictionary["date_extracted"].append(date)
```

## Convert dictionary to pandas dataframe

```python
data = pandas.DataFrame(dictionary)
```
