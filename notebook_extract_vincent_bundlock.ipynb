```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
```

```python
url = 'https://vincentarelbundock.github.io/Rdatasets/datasets.html'
```

```python
response = requests.get(url)
list = []
```

```python
soup = BeautifulSoup(response.content, 'lxml')
```

```python
for link in soup.findAll("a"):
    if link.get("href"):
        list.append(link.get("href"))
```

```python
no_dupes_list = []
for item in list:
    if item not in no_dupes_list:
        no_dupes_list.append(item)
```

```python
no_dupes_list
```

```python
for link in no_dupes_list:
    x = link.split("/")
    name = x[len(x)-1]
    r = requests.get(link, allow_redirects = True)
    open("datasets\\" + name, 'wb').write(r.content)
```
