# Pull data from fixed speed cameras site

```python
import pandas 
import bs4
import requests
import datetime
```

```python
path = "/users/danielcorcoran/Desktop/fixed_camera.html"
```

```python
soup = bs4.BeautifulSoup(open(path), "lxml")
```

```python
print(soup.prettify())
```

```python
geolocations = soup.find_all("div", {"class":"geolocation"})
```

```python
dictionary = {"location_long":[],
             "location_short":[],
             "longitude":[],
             "latitude":[]}
```

```python
for location in geolocations:
    location_long = location.find("div",{"class":"data-row location"}).text.strip()
    latitude = location.get("data-lat")
    longitude = location.get("data-lng")
    location_short = location.find("h2").text.strip()
    
    dictionary["location_long"].append(location_long)
    dictionary["location_short"].append(location_short)
    dictionary["latitude"].append(latitude)
    dictionary["longitude"].append(longitude)
```

```python
dictionary
```

```python
data = pandas.DataFrame(dictionary)
```
