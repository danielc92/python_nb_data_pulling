# Combine files in grants_output folder to one

```python
import os
import pandas
```

```python
input_path = "/Users/danielcorcoran/Desktop/github_repos/python_nb_data_pulling/grant_output/"
```

```python
data_list = []

for filename in os.listdir(input_path):
    print("processing",filename)
    data_list.append(pandas.read_csv(input_path + filename))
```

```python
combined_data = pandas.concat(data_list, axis = 0, sort = True)
```
